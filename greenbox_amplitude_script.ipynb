```python
import polars as pl
```

```python
df_events = pl.read_csv('greenbox_events.csv')
df_spend = pl.read_csv('greenbox_spend.csv')
df_users = pl.read_csv('greenbox_users.csv')
```

# **Creating a Merged Dataset specifically for Amplitude**

```python
# We join User Properties (Channel Device, AB Group) onto every Event
amplitude_df = (
    df_events.join(df_users, on='user_id',how='left')
    .select([
        pl.col('user_id'),
        pl.col('event_name').alias('event_type'), # Amplitude calls it 'event_type'
        pl.col("timestamp").alias('time'), # Amplitude calls it 'time

        #User Properties (The Data I want to filter)
        pl.col('acquisition_channel'),
        pl.col('device'),
        pl.col('ab_test_group'),
        pl.col('revenue')
    ])
)
```

```python
amplitude_df.write_csv('greenbox_amplitude_import.csv')
print('greenbox_amplitude_import.csv was created and ready for download')
```
